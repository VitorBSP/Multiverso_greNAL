```python
cidades['Cidade'] = cidades['Cidade'].str.normalize('NFKC').map(lambda x: x[1:] if x[0] == ' ' else x).map(lambda x: x[:-1] if x[-1] == ' ' else x)
cidades['Mesorregioes'] = cidades['Mesorregioes'].str.normalize('NFKC').map(lambda x: x[1:] if x[0] == ' ' else x).map(lambda x: x[:-1] if x[-1] == ' ' else x)
cidades['Microrregioes'] = cidades['Microrregioes'].str.normalize('NFKC').map(lambda x: x[1:] if x[0] == ' ' else x).map(lambda x: x[:-1] if x[-1] == ' ' else x)
cidades['Populacao_2010'] = cidades['Populacao_2010'].str.normalize('NFKC').str.replace(' ', '').str.replace(',', '').map(lambda x: None if '[3]' in x else x).astype(float).astype('Int64')
cidades['Populacao_2021'] = cidades['Populacao_2021'].str.normalize('NFKC').str.replace(' ', '').str.replace(',', '').astype('int64')
```

```python
pib = pib.drop(pib.index[0])
```

```python
pd.options.mode.chained_assignment = None
pib1 = pib[pib['Posicao'] == '1.1']
pib2 = pib[pib['Posicao'] == '2.1']
pib1['Localidade'] = pib1['Localidade'].str.normalize('NFKC').map(lambda x: x[1:] if x[0] == ' ' else x).map(lambda x: x[:-1] if x[-1] == ' ' else x)
pib2['Localidade'] = pib2['Localidade'].str.normalize('NFKC').map(lambda x: x[1:] if x[0] == ' ' else x).map(lambda x: x[:-1] if x[-1] == ' ' else x)
```

```python
pib1 = pib1.drop(columns=['Posicao', 'Nome', '1999', '2000', '2001','2002','2003', 
                    '2004', '2005', '2006', '2007', '2008', '2009', '2011', '2012', 
                    '2013', '2014', '2015', '2016', '2017', '2018', 'Unidade', 'Multiplicador'])
pib2 = pib2.drop(columns=['Posicao', 'Nome', '1999', '2000', '2001','2002','2003', 
                    '2004', '2005', '2006', '2007', '2008', '2009', '2011', '2012', 
                    '2013', '2014', '2015', '2016', '2017', '2018', 'Unidade', 'Multiplicador'])
```

```python
idh = idh.drop(idh.index[len(idh)-1]).drop(columns=['Posicao', 'Nome', 'Unidade', 'Multiplicador'])
idh['Localidade'] = idh['Localidade'].str.normalize('NFKC').map(lambda x: x[1:] if x[0] == ' ' else x).map(lambda x: x[:-1] if x[-1] == ' ' else x)
```

```python
cidades = cidades .\
    merge(pib1, how='left', left_on='Cidade', right_on = "Localidade").drop(columns=['Localidade']).rename(columns={"2010": "PIB_2010", "2019" : "PIB_2019"}) .\
    merge(pib2, how='left', left_on='Cidade', right_on = "Localidade").drop(columns=['Localidade']).rename(columns={"2010": "PIB_per_capita_2010", "2019" : "PIB_per_capita_2019"}) .\
    merge(idh, how='left', left_on='Cidade', right_on = "Localidade").drop(columns=['Localidade']).rename(columns={"2010": "IDH_2010", "2000" : "IDH_2000","1991" : "IDH_1991"})    
```

```python
cidades['PIB_2010'] = cidades['PIB_2010'].map(lambda x: None if '-' in str(x) else x)
cidades['PIB_per_capita_2010'] = cidades['PIB_per_capita_2010'].map(lambda x: None if '-' in str(x) else x)
cidades['IDH_2010'] = cidades['IDH_2010'].map(lambda x: None if '-' in str(x) else x)
cidades['IDH_2000'] = cidades['IDH_2000'].map(lambda x: None if '-' in str(x) else x)
cidades['IDH_1991'] = cidades['IDH_1991'].map(lambda x: None if '-' in str(x) else x)
```
