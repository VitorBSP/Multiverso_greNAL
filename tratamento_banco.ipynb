```python
import pandas as pd
import numpy as np
import unicodedata
```

```python
campeonatos = pd.read_csv('campeonatos_gauchos.csv', sep=',', encoding='utf-8')
times = pd.read_csv('times_gauchos.csv', sep=',', encoding='utf-8')
```

```python
campeonatos[['Primeiro', 'Numero']] = campeonatos['Primeiro'].str.split(' \(', expand=True)
campeonatos['Primeiro'] = campeonatos['Primeiro'].map(lambda x: x[:x.find('[d]')] if '[d]' in x else x).map(lambda x: x[:x.find('-RS')] if '-RS' in x else x)
#campeonatos.iloc[8,2][:campeonatos.iloc[8,2].find('[d]')] 
#print('[d]' in campeonatos.iloc[8,2], campeonatos.iloc[8,2].find('[d]'),campeonatos.iloc[8,2][:13])
campeonatos['Primeiro'] = campeonatos['Primeiro'].str.normalize('NFKD').str.replace(' ', '', n=1).map(lambda x: x[:-1] if x[-1] == ' ' else x)
campeonatos['Edicao'] = campeonatos['Edicao'].str.extract(r'(\d+)', expand=False)
campeonatos['Ano'] = campeonatos['Ano'].str.extract(r'(\d+)', expand=False)
#campeonatos['Primeiro'] = campeonatos['Primeiro'].str.normalize('NFKD')
times['Apelido'] = times['Apelido'].str.normalize('NFKD')
#campeonatos['Primeiro'] = campeonatos['Primeiro'].str.encode('utf-8')
#times['Apelido'] = times['Apelido'].str.encode('utf-8')
#campeonatos['Ano'] = campeonatos['Ano'].str.replace(r'[(detalhes)]', '')
#campeonatos = campeonatos.drop(columns=['Numero'])
del campeonatos['Numero']
campeonatos.head(5)
```

```python
campeonatos['Segundo'] = campeonatos['Segundo'].map(lambda x: x[:max(map(lambda y: x.find(y) if y in x else 0, ['[a]','[b]', '[c]', '[d]']))] if ('[a]' in x) or ('[b]' in x) or ('[c]' in x) or ('[d]' in x)  else x).map(lambda x: x[:x.find('-RS')] if '-RS' in x else x)
campeonatos['Segundo'] = campeonatos['Segundo'].str.normalize('NFKD').str.replace(' ', '', n=1).map(lambda x: x[:-1] if x[-1] == ' ' else x)
campeonatos['Terceiro'] = campeonatos['Terceiro'].map(lambda x: x[:max(map(lambda y: x.find(y) if y in x else 0, ['[a]','[b]', '[c]', '[d]']))] if ('[a]' in x) or ('[b]' in x) or ('[c]' in x) or ('[d]' in x)  else x).map(lambda x: x[:x.find('-RS')] if '-RS' in x else x)
campeonatos['Terceiro'] = campeonatos['Terceiro'].str.normalize('NFKD').str.replace(' ', '', n=1).map(lambda x: x[:-1] if x[-1] == ' ' else x)
campeonatos['Quarto'] = campeonatos['Quarto'].map(lambda x: x[:max(map(lambda y: x.find(y) if y in x else 0, ['[a]','[b]', '[c]', '[d]']))] if ('[a]' in x) or ('[b]' in x) or ('[c]' in x) or ('[d]' in x)  else x).map(lambda x: x[:x.find('-RS')] if '-RS' in x else x)
campeonatos['Quarto'] = campeonatos['Quarto'].str.normalize('NFKD').str.replace(' ', '', n=1).map(lambda x: x[:-1] if x[-1] == ' ' else x)
#max(list(map(lambda x: a.find(str(x)), ['[a]','[b]', '[c]', '[d]'])))
campeonatos.head(5)
```
